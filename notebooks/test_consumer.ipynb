```python
from confluent_kafka import Consumer
import datetime
import json
import csv
import ast
```

```python
# Configuration for the bootsraping services
conf = {'bootstrap.servers':'broker:9092','group.id':'python-consumer','auto.offset.reset':'earliest'}
consumer = Consumer(conf)
```

```python
topic_name = 'lidarOut' # topic name (has to be the tag for parquet files in ../data/raw)
consumer.subscribe([topic_name])
```

```python
# Consume data
while True:
    msg=consumer.poll(1.0) #timeout
    if msg is None:  
        print('Waiting...')
        continue
    elif msg.error():
        print('Error: {}'.format(msg.error()))
        continue
    else:
        x = msg.value().decode('utf-8').replace('\\n',', ')
        print(x)
        x = ast.literal_eval(x)
            
consumer.close()
```
